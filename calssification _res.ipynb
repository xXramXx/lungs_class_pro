```python
from tensorflow.keras.models import load_model
import matplotlib.pylab as plt
import numpy as np
import cv2

model_2=load_model('F:\\Explore\\mechine\\paper\\pro\\lung_tumor\\weights\\lung_class_own_model_1.h5',compile=False)
```

```python
def shape_data(img):
    img=cv2.resize(img,(256,256))
    img=cv2.normalize(img,None,alpha=0,beta=1,norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    return img 
```

```python
cls=['Bengning_cases', 'Malignant_cases', 'Normal_cases']
img=plt.imread('F:\\Explore\\mechine\\paper\\dataset\\lung\\Normal_cases\\Normal case (22).jpg')
img1=shape_data(img)
img2=np.array(img1).reshape(-1,256,256,3)
pre_dat = model_2.predict(img2)
plt.imshow(img1)
cls[np.argmax(pre_dat)]
```
